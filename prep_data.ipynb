```python
%matplotlib inline
import matplotlib.pyplot as plt
from IPython.display import Image
import os
import numpy as np
from PIL import Image
import cv2
from skimage import io


dir_prefix = '/Users/chitrangtalaviya/Documents/Deep_Learning_Car_295B/'
dirProcessForVehicles = '%s/vehicles/MiddleClose' % (dir_prefix)

dir_prefix = '/Users/chitrangtalaviya/Documents/Deep_Learning_Car_295B/'
dirProcessForNonVehicles = '%s/non-vehicles/MiddleClose' % (dir_prefix)




IMAGE_WIDTH = 64
IMAGE_HEIGHT = 64

cur_dir = os.getcwd()
print("resizing images")
print("current directory:",cur_dir)


rootdir = dirProcessForVehicles
i = 0
for subdir, dirs, files in os.walk(rootdir):
    for file in files:
        print os.path.join(subdir, file)
        #print new_filename 
        #os.rename(os.path.join(subdir, file), new_filename)
        a = np.array(io.imread(os.path.join(subdir, file)))
        b = cv2.resize(a, (IMAGE_WIDTH,IMAGE_HEIGHT), interpolation=cv2.INTER_CUBIC)
        new_filename = "imgVehicle" + str(i) + ".png"
        i = i + 1
        img = Image.fromarray(b, 'RGB')
        img.save(new_filename)
```
